## Pyplot vs Matplotlib OO method
    > for plotting something quickly - use pyplot
    > for plotting something complex - use oo method

```python
over_50 = heart_disease[heart_disease["age"]>50]
```

```python
over_50
```

```python
#OO Method mixed with pyplot method

fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
            x='age',
            y='chol',
            c= 'target',
            ax=ax );

# ax.set_xlim([45,100]);
```

```python
# pyplot method

over_50.plot(kind="scatter",
            x='age',
            y='chol',
             c='target',
            figsize=(10,6)
            );
```

# OO method from scratch

```python
# OO method from the scratch

fig, ax = plt.subplots(figsize = (10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]
                    );

# customize the plot
ax.set(title = "Heart Disease & Cholestrol Level",
      xlabel = 'Age',
      ylabel = 'Cholestrol');

# Add a legend

ax.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = '--' );
```

```python
over_50.head(10)
```

```python
# subplot of age, chol, thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize = (10,10),
                             sharex=True)

# plot on ax0 age vs chol
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                    c=over_50["target"]);


# customize ax0
ax0.set(title="Heart Disease and Cholestrol Level",
#        xlabel="Age", (since x axis is shared among both)
       ylabel='Cholestrol');

# add legend
ax0.legend(*scatter.legend_elements(), title = "Target");

# add horizontal line - a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# plot ax1 age vs thalach
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target']);

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel='Age',
       ylabel='Max Heart Rate');

# add legend
ax1.legend(*scatter.legend_elements(), title = 'Target');

# add horizontal line - a mean line
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle = '--');

# Add a title to the entire fig

fig.suptitle(" Heart Disease Analysis", fontsize=16, fontweight='bold');
```

```python
# set style

plt.style.use('seaborn-whitegrid')

fig, ax = plt.subplots(figsize = (10,6))

# plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap='winter'); #changing color scheme

# customize the plot
ax.set(title = "Heart Disease & Cholestrol Level",
      xlabel = 'Age',
      ylabel = 'Cholestrol');

# Add a legend

ax.legend(*scatter.legend_elements(), title = "Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle = '--' ); 
```

```python
# Heart Disease Analysis Graph Recreating

fig, (ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             figsize = (10,10),
                             sharex=True)

# plot on ax0 age vs chol
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap='winter');


# customize ax0
ax0.set(title="Heart Disease and Cholestrol Level",
#        xlabel="Age", (since x axis is shared among both)
       ylabel='Cholestrol');

# set x axis limit
ax0.set_xlim([50,80]);


# add legend
ax0.legend(*scatter.legend_elements(), title = "Target");

# add horizontal line - a mean line
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--",
           c='red');

# plot ax1 age vs thalach
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                      cmap='winter');

# customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel='Age',
       ylabel='Max Heart Rate');

# add legend
ax1.legend(*scatter.legend_elements(), title = 'Target');

# add horizontal line - a mean line
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle = '--',
           c='red');

# set y axis limit
ax1.set_ylim([60,200]);

# Add a title to the entire fig

fig.suptitle(" Heart Disease Analysis", fontsize=16, fontweight='bold');
```
